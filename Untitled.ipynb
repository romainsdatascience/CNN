```python
valid_split, train_split = tfds.Split.TRAIN.subsplit([15, 75])
test_set = dataset['test']
valid_set = tfds.load("horses_or_humans", split=valid_split, as_supervised=True) 
train_set = tfds.load("horses_or_humans", split=train_split, as_supervised=True)
```

```python
plt.figure(figsize=(12, 10))
index = 0
for image, label in train_set.take(12):
    index += 1
    plt.subplot(3, 4, index)
    plt.imshow(image)
    plt.title("Class: {}".format(class_names[label]))
    plt.axis("off")

plt.show()
```

```python
def preprocess(image, label):
    resized_image = tf.image.resize(image,[224, 224])
    resized_image = resized_image / 256.0
    return resized_image, label

resized_train = train_set.map(preprocess)

plt.figure(figsize=(12, 10))
index = 0
for image, label in train_set.take(12):
    index += 1
    plt.subplot(3, 4, index)
    plt.imshow(image)
    plt.title("Class: {}".format(class_names[label]))
    plt.axis("off")

plt.show()
```

```python
model = keras.models.Sequential([
    keras.layers.Conv2D(filters=64, kernel_size=6, padding = 'same', input_shape=[224, 224, 3]),
    keras.layers.MaxPooling2D(pool_size=2),
    keras.layers.Conv2D(128, 3, activation="relu", padding="same"),
    keras.layers.Conv2D(128, 3, activation="relu", padding="same"),
    keras.layers.MaxPooling2D(pool_size=2),
    keras.layers.Conv2D(256, 3, activation="relu", padding="same"),
    keras.layers.Conv2D(256, 3, activation="relu", padding="same"),
    keras.layers.MaxPooling2D(pool_size=2),
    keras.layers.Flatten(),
    keras.layers.Dense(units=128, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(units=64, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(units=2, activation='softmax'),
])
```

```python
def preprocess(image, label):
    resized_image = tf.image.resize(image,[224, 224])
    resized_image = resized_image / 256.0
    return resized_image, label

resized = train_set.map(preprocess)

index = 0
for image, label in resized.take(1):
    print(image.numpy)
```

```python
batch_size = 32
train_set = train_set.shuffle(1000).repeat()
train_set = train_set.map(preprocess).batch(batch_size).prefetch(1) 
valid_set = valid_set.map(preprocess).batch(batch_size).prefetch(1) 
test_set = test_set.map(preprocess).batch(batch_size).prefetch(1)
```

```python
optimizer = keras.optimizers.SGD(learning_rate=0.0001, momentum=0.9,
                                 nesterov=True, decay=0.001)
optimizer2 = tf.keras.optimizers.Adam(learning_rate = 0.001)
model.compile(loss="binary_crossentropy", optimizer=optimizer2,
              metrics=["accuracy"])
history = model.fit(train_set,
                    steps_per_epoch=int(0.75 * train_size / batch_size),
                    validation_data=valid_set,
                    validation_steps=int(0.15 * train_size / batch_size),
                    epochs= 10)
```
